```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("C:\\Users\\spofa\\Desktop\\HonsISMan\\Learning From Data\\fifaData.xlsx")
print(dataset)
```

```python
dataset.plot()
```

```python
sum = dataset.sum()
print(sum)
```

```python
min = dataset.min()
print(min)
```

```python
max = dataset.max()
print(max)
```

```python
mean = dataset.mean()
print(mean)
```

```python
first_quantile = dataset.quantile(0.25)
print(first_quantile)
plt.plot(first_quantile)
plt.show()
```

```python
median = dataset.median()
print(median)
median.plot(kind='bar')
plt.show()
```

```python
third_quantile = dataset.quantile(0.75)
print(third_quantile)
plt.plot(third_quantile)
plt.show()
```

```python
variance = dataset.var()
print(variance)
variance.plot(kind='bar')
plt.show()
```

```python
standard_dev = dataset.std()
print(standard_dev)
```

```python
dataset["Age"].plot(kind='hist')
```

```python
multiply = 1
values = [8,16,22,12,41]
n = len(values)

for i in values:
    multiply = (multiply) * (i)
    
geometric_mean = (multiply)**(1/n)
print(geometric_mean)
```

```python
kurtosis = dataset.kurtosis()
print(kurtosis)
plt.plot(kurtosis)
plt.show()
```

```python
coefficient_of_variation = lambda x: np.std(x, ddof=1) / dataset.mean() * 100 

print(coefficient_of_variation(dataset))
plt.pie(coefficient_of_variation(dataset))
plt.show()
```
