```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 6))
```

```python
dataset = pd.read_excel("C:\\Users\\spofa\\Desktop\\HonsISMan\\Learning From Data\\fifaData.xlsx")
```

```python
sum = dataset.sum()
print(sum)
```

```python
min = dataset.min()
print(min)
```

```python
max = dataset.max()
print(max)
```

```python
mean = dataset.mean()
print(mean)
```

```python
first_quantile = dataset.quantile(0.25)
print(first_quantile)
fig = plt.figure(figsize=(12, 6))
plt.plot(first_quantile, color='green', linewidth=2, marker='o')
plt.title('First Quantile')
plt.xlabel('Player Attributes')
plt.ylabel('25%')
plt.show()
```

```python
median = dataset.median()
print(median)
median.plot(kind='bar',color='maroon')
plt.title('Mean')
plt.xlabel('Player Attributes')
plt.ylabel('Value')
plt.show()
```

```python
third_quantile = dataset.quantile(0.75)
print(third_quantile)
fig = plt.figure(figsize=(12, 6))
plt.plot(third_quantile, color='Orange', linewidth=2, marker='o')
plt.title('Third Quantile')
plt.xlabel('Player Attributes')
plt.ylabel('75%')
plt.show()
```

```python
variance = dataset.var()
print(variance)
fig = plt.figure(figsize=(12, 6))
variance.plot(kind='bar', color='red')
plt.title('Variance')
plt.xlabel('Player Attributes')
plt.ylabel('Variance per attribute')
plt.show()
```

```python
standard_dev = dataset.std()
print(standard_dev)
```

```python
multiply = 1
values = [8,16,22,12,41]
n = len(values)

for i in values:
    multiply = (multiply) * (i)
    
geometric_mean = (multiply)**(1/n)
print(geometric_mean)
```

```python
kurtosis = dataset.kurtosis()
print(kurtosis)
fig = plt.figure(figsize=(12, 6))
plt.plot(kurtosis,  color='purple', linewidth=2, marker='o')
plt.title('Kurtois')
plt.xlabel('Player Attributes')
plt.ylabel('Value')
plt.show()
```

```python
coefficient_of_variation = lambda x: np.std(x, ddof=1) / dataset.mean() * 100 
labels = ['Club_Kit', 'Rating', 'Age', 'Ball_Control', 'Dribbling', 'Marking', 'Vision', 'Composure', 'Short_Pass', 'Speed', 'Weight']
print(coefficient_of_variation(dataset))
fig = plt.figure(figsize=(12, 10))
plt.pie(coefficient_of_variation(dataset), labels=labels)

plt.show()
```

```python
absolutes = dataset.abs()
print(absolutes)
```

```python
mean_absolute_deviation = (dataset - dataset.mean()).abs().mean()
print(mean_absolute_deviation)
```

```python
cumulative_sum = dataset.cumsum()
print(cumulative_sum)
```

```python
cumulative_product = dataset.cumprod()
print(cumulative_product)
```

```python
#correlation between ball control and dribbling
correlation = dataset['Ball_Control'].corr(dataset['Dribbling'])
dataset.plot.scatter(x='Dribbling', y='Ball_Control')
print(correlation)
plt.show()
```

```python
#correlation between passing and vision
correlation = dataset['Short_Pass'].corr(dataset['Vision'])
dataset.plot.scatter(x='Dribbling', y='Ball_Control')
print(correlation)
plt.show()
```

# FIFA 20 complete player dataset

*Information from data:*

- This is a dataset from FIFA 20, consisting of the attributes of all the players registered in the game. Only the top one hundred highly rated players were selected for this study.

- In the top one hundred, the lowest rated player is rated 84/99 overall, and the youngest of the top one hundred is aged 22. A lot of players above the age of 28 tend to be highly rated because the older they get the more skilled and experienced they are, this is proven by the fact that there is a 39 year old in the 100 rated players, but they also get slower because the younger players seem to have more speed than the old players. 

- Bearing in mind that within the top one hundred, there are various positions (goalkeepers, defenders, midfielders, and attackers), but the average player in the top 100 possesses the ability to dribble and to provide a short pass. 

- There is a positive correlation between players short pass and vision, it comes naturally that if a player has good vision then they are gifted with the ability to detect the right moments to play the pass, therefore also getting a good score for short passes because they are will less errors since the passes are made at the right time.

- There is also a strong positive correlation between ball control and dribbling, dribbling requires gentle touches of the ball towards various directions and sudden turns, players who are good at dribbling also have a great ball control score because there is no way one can successfully dribble without good ball control.
